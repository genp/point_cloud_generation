```python
# Cell 2: Imports
import open3d as o3d
import numpy as np
import laspy
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
```

```python
# Cell 3: Helper - Load .las/.laz file and convert to Open3D point cloud
def load_las_to_open3d(filepath):
    las = laspy.read(filepath)
    points = np.vstack((las.x, las.y, las.z)).transpose()
    
    pcd = o3d.geometry.PointCloud()
    pcd.points = o3d.utility.Vector3dVector(points)
    
    return pcd
```

```python
# Plot with matplotlib (top-down 2D view)
def plot_topdown(pcd_obj):
    pts = np.asarray(pcd_obj.points)
    plt.figure(figsize=(8, 6))
    plt.scatter(pts[:, 0], pts[:, 1], c=pts[:, 2], cmap='viridis', s=0.5)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Top-down 2D view (colored by Z)')
    plt.colorbar(label='Height (Z)')
    plt.axis('equal')
    plt.show()


```

```python
#Load a point cloud
# Choose your input file (PLY, PCD, or LAS/LAZ)
# Uncomment the one you want to use

# Load from .ply
# pcd = o3d.io.read_point_cloud("example.ply")

# Load from .pcd
# pcd = o3d.io.read_point_cloud("example.pcd")

# Load from .las or .laz
pcd = load_las_to_open3d("example.las")

# Show basic info
print(pcd)
print("Number of points:", len(pcd.points))

# Plot point cloud
plot_topdown(pcd)
```
